Next step: combining all the months and years for each police force

```python
import pandas as pd

months = [
    "2024-01","2024-02","2024-03","2024-04","2024-05","2024-06","2024-07",
    "2024-08","2024-09","2024-10","2024-11","2024-12",
    "2025-01","2025-02","2025-03","2025-04","2025-05","2025-06","2025-07",
    "2025-08","2025-09","2025-10","2025-11","2025-12",
    "2026-01"
]

forces = ["city-of-london", "essex", "surrey", "west-midlands"]

df_list = []

for month in months:
    for force in forces:
        file_path = f"data/{month}/{month}-{force}-street.csv"
        try:
            temp = pd.read_csv(file_path)
            temp["force"] = force
            temp["year"] = month[:4]
            df_list.append(temp)
        except:
            pass

```

```python
df = pd.concat(df_list, ignore_index=True)
```

# Drop missing rows: latitude/longitude
<hr style="border-top:1px dashed">

Rows with missing latitude or longitude cannot be mapped or spacially analysed and they also come with the LSOA code and LSOA name meaning removing latitude or longityde would remove the same rows where the LSOA code/name are missing

```python
df = df[df["Latitude"].notna() & df["Longitude"].notna()]
```

```python
df.isna().sum()
```

<a name="anchorInstall" style="position:absolute;"></a>
<hr style="border:2px solid">

# Filling Missing Data: Last outcome category
<hr style="border-top:1px dashed">

Sample data shows outcomes however the missing values can mean the case is still pending or not updated which is normal and outcomes can take months to be completed meaning the case could still be open so instead the missing values should be filled with outcome not recorded yet.

```python
df.loc[:, "Last outcome category"] = df["Last outcome category"].fillna("Outcome not recorded yet")

```

```python
df.isna().sum()
```

```python
df['Last outcome category'].unique()
```

<a name="anchorInstall" style="position:absolute;"></a>
<hr style="border:2px solid">

# Filling Missing Data: Crime ID
<hr style="border-top:1px dashed">

Crime ID has a lot of missing values but many crimes can be anonymised or withheld by polce forces meaning missing values does not mean the data is unreliable but it means simply the ID wasn't provided therefore it should be filled.

```python
df.loc[:, "Crime ID"] = df["Crime ID"].fillna("No Crime ID")
```

```python
df["Crime ID"].isna().sum()
```

```python
df.isna().sum()
```

# Population Data
<hr style="border-top:1px dashed">

As the population differs for each force it would not be fair to compare based on the total value so a population data of police force was used to make it a more fair comparison, so a new dataset was downloaded and using excel a new population column was created by suming the rows. The year was also filtered to 2024 to only compare the latest population for a more accurate comparison between forces

```python
pop = pd.read_csv("policeforceareas2024.csv")
pop
```

```python
pop.columns
```

The table contains many columns not required so everything is dropped except the PFA 2023 Name and population

```python
pop=pop[['PFA 2023 Name' , 'population']]
pop
```

The crime data set uses force as the column name so to make it more consistent the PFA 2023 Name is renamed to force

```python
pop = pop.rename(columns = {'PFA 2023 Name' : 'force'})
```

```python
pop
```

The police force names now need to be standardised to match the names of the crime data.

```python
print(pop['force'].unique())
print(df['force'].unique())
```

You can see the population data uses capitals and has no dashes in between spaces

```python
pop["force"] = (
    pop["force"]
    .str.lower()
    .str.replace(" ", "-")
)
```

```python
pop['force'].unique()
```

```python
pop.dtypes
```

The Population data is stored as a string so now that needs to be converted into a number as it is a numeric value

```python
pop['population']= (
    pop['population']
    .astype(str)
    .str.replace(",","")
    .astype(float)
)
```

```python
pop["population"] = pop["population"].astype(int)
pop.dtypes
```

This will now be saved to be used for the analysis notebook

```python
pop.to_csv("population_cleaned.csv", index=False)
```
