Next step: combining all the months and years for each police force

```python
import pandas as pd

months = [
    "2024-01","2024-02","2024-03","2024-04","2024-05","2024-06","2024-07",
    "2024-08","2024-09","2024-10","2024-11","2024-12",
    "2025-01","2025-02","2025-03","2025-04","2025-05","2025-06","2025-07",
    "2025-08","2025-09","2025-10","2025-11","2025-12",
    "2026-01"
]

forces = ["city-of-london", "essex", "surrey", "west-midlands"]

df_list = []

for month in months:
    for force in forces:
        file_path = f"data/{month}/{month}-{force}-street.csv"
        try:
            temp = pd.read_csv(file_path)
            temp["force"] = force
            temp["year"] = month[:4]
            df_list.append(temp)
        except:
            pass

```

```python
df = pd.concat(df_list, ignore_index=True)
```

# Drop missing rows: latitude/longitude
<hr style="border-top:1px dashed">

Rows with missing latitude or longitude cannot be mapped or spacially analysed and they also come with the LSOA code and LSOA name meaning removing latitude or longityde would remove the same rows where the LSOA code/name are missing

```python
df = df[df["Latitude"].notna() & df["Longitude"].notna()]
```

```python
df.isna().sum()
```

# Filling Missing Data: Last outcome category
<hr style="border-top:1px dashed">

Sample data shows outcomes however the missing values can mean the case is still pending or not updated which is normal and outcomes can take months to be completed meaning the case could still be open so instead the missing values should be filled with outcome not recorded yet.

```python
df.loc[:, "Last outcome category"] = df["Last outcome category"].fillna("Outcome not recorded yet")

```

```python
df.isna().sum()
```

```python
df['Last outcome category'].unique()
```

# Filling Missing Data: Crime ID
<hr style="border-top:1px dashed">

Crime ID has a lot of missing values but many crimes can be anonymised or withheld by polce forces meaning missing values does not mean the data is unreliable but it means simply the ID wasn't provided therefore it should be filled.

```python
df.loc[:, "Crime ID"] = df["Crime ID"].fillna("No Crime ID")
```
